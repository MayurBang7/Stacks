# <font color='blue'> Stacks </font>

A stack is a basic data structure that can be logically thought of as a linear structure represented by a real physical stack or pile, a structure where insertion and deletion of items takes place at one end called top of the stack. 


The basic implementation of a stack is also called a LIFO (Last In First Out) to demonstrate the way it accesses data, since as we will see there are various variations of stack implementations.

There are basically three operations that can be performed on stacks. These are 

    1) inserting an item into a stack (push). 
    2) deleting an item from the stack (pop). 
    3) displaying the contents of the stack (peek or top).
    
## <font color='blue'> Implementation of Stacks </font>

As we know based on the memory we can divide our data structure in two types, **static** and **dynamic** data structure. 

**Static** data structure come with limited size and resizing the space is an overhead once the program execution starts. 

Where as **Dynamic** data structures are the once where memory allocation and deallocation happens dynamically as per the requirement of a program.

Lets have a look at the type of implementation we can have for Stack data structure. 

A stack of an unlimited capacity is a dynamic data structure, regardless of its implementation. 

It could be implemented using a linked list or an array which is re-allocated once its capacity is reached. The size of such a stack changes as you add or remove data.

### <font color='blue'> Implementation Using Linked List </font>

Lets explore the implementation of stack using linked list. Before we will see the code lets have a look at the properties and some generic variables that need to be there.

As we know that every linked list can have its own head and tail pointers to point the beginning and the end of the linked list respectively. 

However for the implementation of stack we do not require tail pointer. 

According to stack properties we have access to one variable named as top. This variable will be used to implement the methods in stack class. 

Considering this we can design our head and linked list in such a way that insertion and deletion of nodes in stack will happen using head node only.  

In this section, we are going to us properties of singly linked list where every node will contain at least two information; 
    
    * Node value
    * Reference to the next pointer
    

Lets have a look at the <class Node> and the members of it.         
  

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None
```

Here we have declared a class that has two attributes *data* and *next.* 

- **\<data\>** holds a values assigned by a user
- **\<next\>** holds the address / memory reference of the next node

Lets have a look at the implementation of the stack data structure using linked list. In the implementation we will be implementing methods to insert, delete data in the stack. 

Insert and delete operations are widely known as push and pop operations. 

While inserting and deleting the data elements in stack as mentioned earlier we need to maintain and re-configure the head node after every operation.

## <font color='blue'> Stack Permutation Problem  </font>

Lets discuss another problem statment using Stack. This problem is famously known as Stack permutations problem.

The problem âStack Permutations (Check if an array is stack permutation of other)â states that you are given two lists a[ ] and b[ ] of size n. 

All the elements in the lists should be unique. Write a function to check if the given list b[ ] is the stack permutation of given list a[ ] or not.

Here are the rules that should be follwoed: 

  a. Only delete operations on both lists. 

  b. Use stack operations such as push and pop data from stack.
  
  c. Stack and input list must be empty at the end.


### <font color='blue'> Algorithms  </font>

1. Continuously delete elements from the input list. 
2. Check if it is deleted element is equal to the top of output list or not, 
3. if it is not equal to the top of output queue then we will push the element to stack. 
4. Once we find an element in input list such the top of input list is equal to top of output list, 
5. Delete a single element from both input and output queues, and compare the top of stack and top of output list. 
6. If top of both stack and output lists are equal then delete element from both stack and output list. If not equal, go to step 1. 
6. Repeat 1 to 6 until the input list becomes empty. 
7. If input list and stack are empty then the input list is permutable to output list. 



```python
class Stack: 
# Implementation of Stack using Arrays
# Initialize the Stack with default capacity or given capacity
    def __init__(self):
        self.objects = []
        self.top = -1

    def size(self):                    #Return the current stack size
        return (self.top + 1)

    def isEmpty(self):                 #Return true iff the stack is empty
        return (self.top < 0)

    def push(self, value):               #Push a new element on the stack
        self.top += 1
        self.objects.append(value)

    def peek(self):                      #Return the top stack element
        if (self.isEmpty()):
            print("Stack is empty.")
            return None
        return self.objects[self.top]

    def pop(self):                      #Pop off the stack element
        if (self.isEmpty()):
            print("Stack is Empty."); 
            return 
        elem  = self.objects.pop()
        self.top = self.top-1           #decrement top 
        return elem

def StackPermut(ip, op, n):
    
    tempStack = Stack()
    while (len(ip) != 0):
        ele = ip[0]
        del ip[0]
        if (ele == op[0]):
            del op[0]
            while ((tempStack.size()) != 0):
    
                if (tempStack.peek() == op[0]):
                    tempStack.pop()
                    del op[0]
                else:
                    break
        else:
            tempStack.push(ele)
    
    return (not len(ip) and not (tempStack.size()))
    
# Driver Code
if __name__ == '__main__':

    # Input Queue
    Input = [1, 2, 3]

    # Output Queue
    output = [2, 1, 3]

    n = 3
    if (StackPermut(Input, output, n)):
        print("Yes")
    else:
        print("Not Possible")
```

## <font color='blue'> Approach </font>

We are going to use Stack data structure to implement the feature mentioned above. Based on the added problem above lets look at the design for the solution; 


- A user will generally start with home page over a tab that is empty. 
- To solve the given problem we need to maintain two stacks, BACK_STACK and FORWARD_STACK, both initially empty
- When user visits a new web URL, or hit the Forward button, the current_url should be moved to the BACK_STACK that is being maintained. 
- When user hits Back button or call back method, the existing page URL should be moved to the FORWARD_STACK

Below we have a Stack class designed with methods in it that will help us in implementing and designing the problem statement. 

Lets have a look at the designed class and its methods below.


```python
class Stack: 
# Implementation of Stack using Arrays
# Initialize the Stack with default capacity or given capacity
    def __init__(self):
        self.objects = []
        self.top = -1

    def size(self):                    #Return the current stack size
        return (self.top + 1)

    def isEmpty(self):                 #Return true iff the stack is empty
        return (self.top < 0)

    def push(self, value):               #Push a new element on the stack
        self.top += 1
        self.objects.append(value)

    def peek(self):                      #Return the top stack element
        if (self.isEmpty()):
            print("Stack is empty.")
            return None
        return self.objects[self.top]

    def pop(self):                      #Pop off the stack element
        if (self.isEmpty()):
            print("Stack is Empty."); 
            return 
        elem  = self.objects.pop()
        self.top = self.top-1           #decrement top 
        return elem
```
